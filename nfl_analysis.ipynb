```python
import nfl_data_py as nfl
import pandas as pd
import os
import urllib.request
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnnotationBbox, OffsetImage
```

```python
pbp = nfl.import_pbp_data([2022])
```

##**Data Cleaning**

```python
# Filter the play-by-play data to include only rows with pass or rush plays.
pbp_rp = pbp[(pbp['pass'] == 1) | (pbp['rush'] == 1)]

# Drop rows with missing values in the 'epa', 'posteam', or 'defteam' columns.
pbp_rp = pbp_rp.dropna(subset=['epa', 'posteam', 'defteam'])

# Check the dimensions (number of rows, number of columns) of the resulting DataFrame.
pbp_rp.shape
```

##**Getting Stats**

```python
# Filter play-by-by data to include only pass plays, group data by team and calculate the mean of 'epa' for pass plays.
pass_epa = pbp_rp[(pbp_rp['pass'] == 1)].groupby('posteam')['epa'].mean().reset_index().rename(columns = {'epa' : 'pass_epa'})

# Sort the DataFrame by 'pass_epa' in descending order to find teams with the highest average 'epa' for pass plays.
pass_epa.sort_values('pass_epa', ascending = False)
```

```python
# MERGING THE DATA
# Filter the filtered play-by-play data to include only rush plays, group data by team, and finding mean.
rush_epa = pbp_rp[(pbp_rp['rush'] == 1)].groupby('posteam')['epa'].mean().reset_index().rename(columns = {'epa' : 'rush_epa'})

# Merge the DataFrames 'pass_epa' and 'rush_epa' on the 'posteam' column to combine pass and rush EPA data.
epa = pd.merge(pass_epa, rush_epa, on = 'posteam')

epa.head()
```

```python
# LOADING THE LOGOS
logos = nfl.import_team_desc()[['team_abbr', 'team_logo_espn']]
logos.head()
```

```python
# Initialize empty lists to store logo paths and team abbreviations.
logo_paths = []
team_abbr = []

# Check if the 'logos' directory exists; if not, create it.
if not os.path.exists("logos"):
    os.makedirs("logos")

# Loop through the 'logos' DataFrame and download team logos, storing paths and abbreviations.
for team in range(len(logos)):
    urllib.request.urlretrieve(logos['team_logo_espn'][team], f"logos/{logos['team_abbr'][team]}.tif")  # Download team logos and save them in the 'logos' directory with the abbreviation as the file name.
    logo_paths.append(f"logos/{logos['team_abbr'][team]}.tif")     # Store the logo paths and team abbreviations in respective lists.
    team_abbr.append(logos['team_abbr'][team])

# Create a DataFrame with team abbreviations and corresponding logo file paths.
data = {'team_abbr' : team_abbr, 'logo_path' : logo_paths}
logo_data = pd.DataFrame(data)

# Merge the 'epa' DataFrame with the 'logo_data' DataFrame based on team abbreviations.
epa_with_logos = pd.merge(epa, logo_data, left_on = 'posteam', right_on = 'team_abbr')
```

```python
# CREATING THE GRAPH
# Set the default figure size and enable automatic layout adjustment
plt.rcParams["figure.figsize"] = [10, 7]
plt.rcParams["figure.autolayout"] = True

# function to get an OffsetImage object from the specified image path.
def getImage(path):
    return OffsetImage(plt.imread(path, format="tif"), zoom=.1)

# Extract pass EPA, rush EPA, and logo paths from the 'epa_with_logos' DataFrame.
x = epa_with_logos['pass_epa']
y = epa_with_logos['rush_epa']
paths = epa_with_logos['logo_path']

# Create a new figure and axis for the scatter plot with team logos.
fig, ax = plt.subplots()

# Iterate through pass EPA, rush EPA, and logo paths to add team logos to the scatter plot.
for x0, y0, path in zip(x, y, paths):
   ab = AnnotationBbox(getImage(path), (x0, y0), frameon=False)
   ax.add_artist(ab)

# Set the x and y axis limits to focus on the relevant data points.
plt.xlim(-0.2, 0.3)
plt.ylim(-0.25, 0.15)

# Set plot title, x-axis label, and y-axis label
plt.title("Rush and Pass EPA, 2022")
plt.xlabel("EPA/Pass")
plt.ylabel("EPA/Rush")

# Add reference lines to the scatter plot.
ax.axhline(0, color='gray', linestyle='--', linewidth=0.5)  # Horizontal line at y=0
ax.axvline(0, color='gray', linestyle='--', linewidth=0.5)  # Vertical line at x=0


# Display the scatter plot with team logos.
plt.show()
```

```python
# Filter the play-by-play data to include only pass plays and group by receiver player names.
receiver_yac = pbp_rp[(pbp_rp['pass'] == 1)].groupby('receiver_player_name').agg({'pass': 'count',
        'yards_after_catch': 'sum'}).reset_index().rename(columns = {'pass' : 'targets', 'yards_after_catch' : 'yac'})

# Filter receivers with less than 100 targets, focusing on more significant data.
receiver_yac = receiver_yac[(receiver_yac['targets'] >= 100)]

# Sort the DataFrame by 'yac' (yards after catch) in ascending order.
receiver_yac.sort_values('yac', inplace = True)
```

```python
# Extract receiver names and yards after catch (YAC) data from the 'receiver_yac' DataFrame.
name = receiver_yac['receiver_player_name']
yac = receiver_yac['yac']

# Create a horizontal bar plot to visualize YAC for each receiver in the 2022 season.
plt.figure(figsize=(10, 7))
plt.barh(name, yac, color='skyblue')  # Set a specific color for the bars

# Set plot title and labels
plt.title("Yards After Catch, 2022")
plt.xlabel("YAC")
plt.ylabel("Receiver")

# Display the horizontal bar plot showing YAC for each receiver.
plt.show()
```
